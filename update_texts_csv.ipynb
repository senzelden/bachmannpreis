```python
import pandas as pd
```

```python
texts = pd.read_csv("texts.csv")
texts.rename(columns={"Unnamed: 0": 'text_id'}, inplace=True)
# texts.set_index('text_id', inplace=True)
texts.head()
```

```python
critics = pd.read_csv("critics_all.csv")
critics.tail(30)
```

```python
old_critics = pd.read_csv("critics.csv")
old_critics.head()
```

```python
full_names = []
for i in range(len(old_critics)):
    full_name = ""
    full_name += old_critics.first_name.values[i]
    if type(old_critics.middle_name.values[i]) == str:
        full_name += " " + old_critics.middle_name.values[i]
    else:
        pass
    if type(old_critics.last_name.values[i]) == str:
        full_name += " " + old_critics.last_name.values[i]
    else:
        pass
    full_names.append(full_name)
```

```python
full_names[24] = "Hildegard Elisabeth Keller"
```

```python
full_names[20] = "Ijoma Alexander Mangold"
```

```python
old_critics["full_names"] = full_names
```

```python
old_critics
```

```python
combined_critics = pd.merge(critics, old_critics[['id', 'full_names']], left_on='name', right_on='full_names').drop(columns='full_names').rename(columns={'id': 'old_critic_id'}).copy()
combined_critics.old_critic_id = combined_critics.old_critic_id.astype(float).copy()
combined_critics
```

```python
texts_updated = pd.merge(texts, combined_critics[['critic_id', 'old_critic_id']], how='left', left_on='invited_by', right_on='old_critic_id').drop(columns=['old_critic_id', 'invited_by', 'discussion_summary', 'is_performance']).rename(columns={'critic_id': 'invited_by'}).copy()
texts_updated
```

```python
authors = pd.read_csv('authors.csv')
```

```python
texts_updated = pd.merge(texts_updated, authors[['author_id', 'first_name', 'middle_name', 'last_name']], how='left', left_on='author_id', right_on='author_id').copy()
```

```python
texts_updated
```

```python
pseudo_ids = []
for i in range(len(texts_updated)):
    full_name = ""
    full_name += texts_updated.first_name.values[i]
    if type(texts_updated.middle_name.values[i]) == str:
        full_name += texts_updated.middle_name.values[i]
    else:
        pass
    if type(texts_updated.last_name.values[i]) == str:
        full_name += texts_updated.last_name.values[i]
    else:
        pass
    full_name += str(texts_updated.date_participation.values[i][:4])
    pseudo_ids.append(full_name)
pseudo_ids
```

```python
texts_updated['pseudo_id'] = pseudo_ids
```

```python
passig_texts = pd.read_csv("other_bachmann_data/Klagenfurt-Statistik - Texte.csv")
passig_texts.head()
```

```python
texts_final = pd.merge(texts_updated, passig_texts[['ID', 'Shortlist?', 'Schriftart']], how='left', left_on='pseudo_id', right_on='ID').drop(columns=['first_name', 'middle_name', 'last_name', 'pseudo_id', 'ID', 'on_shortlist']).rename(columns={'Shortlist?': 'on_shortlist', 'Schriftart': 'font'}).copy()
texts_final
```

```python
texts_final['on_shortlist'] = texts_final['on_shortlist'].replace({None: False, 'Shortlist': True}).copy()
```

```python
texts_final['invited_by'] = texts_final.invited_by.astype(pd.Int32Dtype())
```

```python
texts_final
```

```python
passig_authors = pd.read_csv("other_bachmann_data/Klagenfurt-Statistik - Autor innen.csv")
passig_authors.head()
```

```python
texts_final['pseudo_id'] = pseudo_ids
```

```python
texts_complete = pd.merge(texts_final, passig_authors[['ID', 'Leseplatz', 'Lesetag', 'Lesezeitpunkt', 'Lesetage im Jahr der Teilnahme']], how='left', left_on='pseudo_id', right_on='ID').drop(columns=['day_reading', 'pseudo_id', 'ID', 'full_name']).rename(columns={'Leseplatz': 'reading_order', 'Lesetag': 'day_reading', 'Lesezeitpunkt': 'reading_context', 'Lesetage im Jahr der Teilnahme': 'total_days_reading'}).copy()
texts_complete
```

```python
texts_complete['reading_order'] = texts_complete.reading_order.astype(pd.Int32Dtype())
texts_complete['total_days_reading'] = texts_complete.total_days_reading.astype(pd.Int32Dtype())
```

```python
texts_complete['reading_context'].unique()
```

```python
texts_complete['reading_context'] = texts_complete['reading_context'].replace({"LT": "last_reading", "ET": "first_reading", "VP": "before_break", "NP": "after_break", "LL": "last_reading"}).copy()
```
