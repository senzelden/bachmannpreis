```python
old_critics = pd.read_csv("critics.csv")
old_critics.head()
```

```python
old_critics["full_name"]
```

```python
old_critics["full_name"] = old_critics["first_name"].str.cat(old_critics[["middle_name", "last_name"]].fillna("").astype(str), sep=" ").str.replace("  ", " ")
old_critics.head()
```

```python
all_critics = pd.read_csv("critics_all.csv")
```

```python
all_critics.head()
```

```python
critics_updated = pd.merge(all_critics, old_critics[['full_name', 'birth_place', 'birth_date', 'gender', 'country', 'works_as_critic', 'is_literary_scholar', 'is_writer']], 'left', left_on='name', right_on='full_name').drop(columns=['name', 'full_name'])
critics_updated
```

```python
critics_updated.set_index('critic_id').to_csv("critics_updated.csv")
```
