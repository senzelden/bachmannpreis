```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
response = requests.get("https://bachmannpreis.orf.at/stories/3047191/")
```

```python
page = soup(response.text, "html.parser")
```

```python
critics = {}
```

```python
critics[str(1977)] = {}
```

```python
critics[str(1977)]["0"]["name"] = "Rolf Becker"
```

```python
critics[str(1977)]["name"] = "Humbert Fink"
```

```python
critics
```

```python
critics = []
for i, lists in enumerate(page.find_all("ul")[4:-5]):
    print(i + 1977)
    if (i + 1977) <= 2003:
        for j in lists.find_all("li"):
            critic = [(i + 1977)] + j.text.split(", ")
            print(critic)
            critics.append(critic)
    else:
        for k in lists.find_all("li"):
            critic = [(i + 1977)] + k.text.strip(")").split(" (")
            print(critic)
            critics.append(critic)
```

```python
critics
```

```python
import pandas as pd


df = pd.DataFrame(critics)
```

```python
df.rename(columns={0: 'year', 1: 'name', 2: 'living_place', 3: 'is_chairman'}, inplace=True)
```

```python
df.name = pd.Categorical(df.name)
```

```python
df['critic_id'] = df.name.cat.codes
```

```python
single_critics = df.groupby('name')[['year']].min().sort_values(['year'])
```

```python
new_df = pd.read_csv('jury_members_over_time.csv')
new_df.drop(columns="Unnamed: 0", inplace=True)
new_df
```

```python
single_critics = new_df.groupby('name')[['year']].min().sort_values(['year'])
names_lists = single_critics.index.str.split()
```

```python
last_names = [full_name[-1] for full_name in names_lists]
```

```python
first_names = [full_name[0] for full_name in names_lists]
middle_names = [full_name[1] for full_name in names_lists if full_name[1] is not full_name[-1]]
```

```python
middle_names = []
for full_name in names_lists:
    if full_name[1] is not full_name[-1]:
        middle_names.append(full_name[1])
    else:
        middle_names.append(None)
```

```python
middle_names
```

```python
single_critics['first_name'] = first_names
single_critics['middle_name'] = middle_names
single_critics['last_name'] = last_names

len(set(single_critics.index.values))
```

```python
critics_ids = single_critics.sort_values(['year', 'last_name']).reset_index().reset_index().rename(columns={'index': 'critic_id'}).set_index('critic_id').drop(columns='year')
critics_ids
```

```python
critics_ids.to_csv('critics_all.csv')
```

```python
critics_ids.reset_index()
```

```python
critics_participations = pd.merge(new_df, critics_ids.reset_index()[['critic_id', 'name']], on='name')
critics_participations
```

```python
critics_participations = critics_participations.sort_values(['year', 'critic_id']).set_index('year').copy()
```
