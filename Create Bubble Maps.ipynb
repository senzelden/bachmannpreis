```python
import os
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base
from credentials import PG_USER, PG_PASS, PG_HOST
```

```python
USERNAME = PG_USER
PASSWORD = PG_PASS # os.getenv("PG_PASS")
HOST = PG_HOST
PORT = '5432'
DATABASE = 'bachmann'

conn_string = f'postgres://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}'
engine = create_engine(conn_string, encoding='latin1', echo=True)
Base = declarative_base(bind=engine)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
class Authors(Base):
    
    __tablename__ = 'authors'
    __table_args__ = {'autoload': True}
    

class Texts(Base):

    __tablename__ = 'texts'
    __table_args__ = {'autoload': True}
    

class Authors_Cities(Base):
    
    __tablename__ = 'authors_cities'
    __table_args__ = {'autoload': True}
    
    
class Authors_Countries(Base):
    
    __tablename__ = 'authors_countries'
    __table_args__ = {'autoload': True}
    
    
class Cities_Geodata(Base):
    
    __tablename__ = 'cities_geodata'
    __table_args__ = {'autoload': True}
    
    
class Countries_Geodata(Base):
    
    __tablename__ = 'countries_geodata'
    __table_args__ = {'autoload': True}
```

```python
from sqlalchemy import func


session.query(func.count(Authors_Cities.city_id), Cities_Geodata.city_name).filter(Texts.author_id==Authors.author_id).filter(Texts.text_id==Authors_Cities.text_id==Authors_Countries.text_id).filter(Authors_Cities.city_id==Cities_Geodata.city_id).filter(Authors_Countries.country_id==Countries_Geodata.country_id).group_by(Authors_Cities.city_name).all()
```
