```python
import pandas as pd
```

```python
texts = pd.read_csv('texts_updated.csv')
authors = pd.read_csv('authors_final.csv')
authors_cities = pd.read_csv('cities_updated.csv')
authors_countries = pd.read_csv('countries_updated.csv')
sentiments = pd.read_csv('summary_sentiments.csv')
text_infos = pd.read_csv('text_infos_updated.csv')
results = pd.read_csv('results.csv')
prices = pd.read_csv('prices.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


y = df9["price_won"]
X = df9.drop(columns='price_won')
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
rf_model.fit(X_train, y_train)
rf_model.score(X_test, y_test)
```

```python
importance_features = pd.DataFrame({'feature':X_train.columns, 'importance': rf_model.feature_importances_}).sort_values(by='importance',ascending=False)
importance_features
```
