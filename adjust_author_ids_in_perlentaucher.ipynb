```python
import pandas as pd
```

```python
perlentaucher = pd.read_csv("perlen_2020.csv")
```

```python
perlentaucher
```

```python
authors = pd.read_csv("authors_wo_duplicates.csv")
authors = authors.reset_index().rename(columns={'index': 'author_id_new'}).copy()
```

```python
perlentaucher_updated = pd.merge(perlentaucher, authors[['author_id', 'author_id_new']], "left", left_on="author_id", right_on="author_id").drop(columns='author_id').rename(columns={'author_id_new': 'author_id'})
perlentaucher_updated.author_id = perlentaucher_updated.author_id.astype(pd.Int32Dtype()).copy()
perlentaucher_updated
```

```python
perlentaucher_updated.columns
```

```python
perlentaucher_updated = perlentaucher_updated[['author_id', 'book_title', 'publisher', 'release_year', 'perlentaucher_url']]
perlentaucher_updated.set_index('author_id')
```

```python
perlentaucher_updated.set_index('author_id').to_csv("perlentaucher_updated.csv")
```
