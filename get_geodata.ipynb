```python
import tqdm
import time
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="Nothing_specific")
```

```python
location = geolocator.geocode("Verden/Aller")
```

```python
print((location.latitude, location.longitude))
```

```python
cities = pd.read_csv("cities.csv")
```

```python
authors = pd.read_csv("authors_final.csv")
```

```python
birth_places = authors.birth_place.unique().tolist()
```

```python
living_places = cities.city.unique().tolist()
```

```python
places = birth_places + living_places
```

```python
places = set(places)
places = list(places)
```

```python
cleaned_places = [x for x in places if str(x) != 'nan']

cleaned_places.sort()

cleaned_places
```

```python
latitudes = []
longitudes = []
for i, place in tqdm.tqdm(enumerate(cleaned_places)):
    time.sleep(1)
    location = geolocator.geocode(place)
    if location != None:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
    else:
        latitudes.append(np.nan)
        longitudes.append(np.nan)
```

```python
cities_geodata = pd.DataFrame({"city": cleaned_places, "lat": latitudes, "lon": longitudes, "distance": distances})
```

```python
cities_geodata
```

```python
cities_geodata.to_csv("cities_geodata.csv")
```
