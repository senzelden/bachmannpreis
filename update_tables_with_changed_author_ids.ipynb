```python
import pandas as pd
```

```python
authors_wo = pd.read_csv("authors_wo_duplicates.csv")
```

```python
authors_wo.reset_index(inplace=True)
```

```python
authors_wo.rename(columns={'index':'author_id_new'}, inplace=True)
```

```python
authors_wo
```

```python
authors_wo["full_name"] = authors_wo["first_name"].str.cat(authors_wo[["middle_name", "last_name"]].fillna("").astype(str), sep="") + "2020"
authors_wo.head()
```

```python
passig_authors = pd.read_csv("other_bachmann_data/Klagenfurt-Statistik - Autor innen.csv")
```

```python
passig_authors
```

```python
authors_updated = pd.merge(authors_wo, passig_authors[['ID', 'Geburtsort']], 'left', left_on='full_name', right_on='ID').drop(columns=['birth_place', 'ID', 'full_name']).rename(columns={'Geburtsort': 'birth_place'})
```

```python
authors_updated
```

```python
authors_updated.columns
```

```python
authors_updated = authors_updated[['author_id_new', 'author_id', 'first_name', 'middle_name', 'last_name',
       'birth_date', 'birth_place', 'gender', 'has_wikipedia', 'website', 'literaturinstitut',
       'has_phd']]
authors_updated
```

```python
authors_updated[authors_updated.last_name == "Helfer"]
```

```python
texts = pd.read_csv("texts_complete.csv")
```

```python
texts_updated = pd.merge(texts, authors_updated[['author_id', 'author_id_new']], 'left', left_on='author_id', right_on='author_id').drop(columns=['author_id']).rename(columns={'author_id_new': 'author_id'})
```

```python
texts_updated.author_id = texts_updated.author_id.astype(pd.Int32Dtype())
texts_updated.reading_order = texts_updated.reading_order.astype(pd.Int32Dtype())
texts_updated.day_reading = texts_updated.day_reading.astype(pd.Int32Dtype())
texts_updated.total_days_reading = texts_updated.total_days_reading.astype(pd.Int32Dtype())
texts_updated.invited_by = texts_updated.invited_by.astype(pd.Int32Dtype())
```

```python
texts_updated
```

```python
texts_updated.columns
```

```python
texts_updated = texts_updated[['text_id', 'author_id', 'text_title', 'date_participation', 'invited_by',
       'on_shortlist', 'font', 'reading_order', 'day_reading',
       'reading_context', 'total_days_reading']]
```

```python
texts_updated.set_index('text_id').to_csv("texts_updated.csv")
```

```python
goodreads = pd.read_csv("goodreads_results_all.csv")
goodreads
```

```python
goodreads_updated = pd.merge(goodreads, authors_updated[['author_id', 'author_id_new']], 'left', left_on='author_id', right_on='author_id').drop(columns=['author_id']).rename(columns={'author_id_new': 'author_id'})
goodreads_updated.author_id = goodreads_updated.author_id.astype(pd.Int32Dtype())
goodreads_updated
```
