```python
import pandas as pd
```

```python
authors_cities = pd.read_csv("cities.csv")
authors_cities
```

```python
cities_geodata = pd.read_csv("cities_geodata.csv")
cities_geodata
```

```python
new_cities = pd.merge(authors_cities, cities_geodata[["city_id", 'city_name']], "left", left_on="city", right_on="city_name").drop(columns='city_name').set_index('text_id')
```

```python
new_cities['city_id'] = new_cities['city_id'].astype(pd.Int32Dtype()).copy()
```

```python
new_cities.to_csv('cities_updated.csv')
```

```python
new_cities.reset_index().to_csv('cities_updated.csv')
```

```python
authors_countries = pd.read_csv("countries.csv")
```

```python
authors_countries.to_csv("countries_updated.csv")
```

```python
unique_countries = authors_countries['author_country'].unique()
```

```python
unique_df = pd.DataFrame(unique_countries)
unique_df.to_csv('unique_countries.csv')
```

```python
unique_df
```

```python
countries_updated = pd.merge(authors_countries,unique_df.reset_index(), "left", left_on="author_country", right_on=0).drop(columns=0).rename(columns={"index_x": "text_id", "index_y": "country_id"})
```

```python
countries_updated.to_csv("countries_updated.csv")
```
